splitting the data using stratified splitting

```python
c_features_training ,c_features_testing,c_target_training,c_target_testing=train_test_split(c_features,c_target,test_size=0.20,stratify=c_target,random_state=42)
```

scaler.fit_transform vs scaler.transform

fit() (Learn): The teacher looks at the exam scores and calculates the Class Average (Mean) and Spread (Standard Deviation). No grades are changed yet; the teacher is just learning the stats.

transform() (Apply): The teacher takes a specific score and compares it to the Class Average to give it a "Grade" (Z-Score).

fit_transform() (Do Both): The teacher calculates the average AND grades the papers in one step.

scaler.fit_transform(X_train)
Logic: "Calculate the average of this training set, and then scale these numbers based on that average."

```python
scaler=StandardScaler()
#getting column names to create dataframe later
cols=c_features_training.columns
#applying the scaling maths to the feature(average and step deviation)
c_features_training=scaler.fit_transform(c_features_training)
#transforming the test data based on the average
c_features_testing=scaler.transform(c_features_testing)
#reconverting to dataframe
c_features_training=pd.DataFrame(c_features_training,columns=cols)
c_features_testing=pd.DataFrame(c_features_testing,columns=cols)
#dumping the scaled features in joblib(binary file)
joblib.dump(scaler,'scaler.joblibv1')
print("scaling complete,features scaled")
print(f"current shape:{c_features_training.shape}")
```

```python
#concatinating the features and target test data to make a complete test dataset file
# 1. Reset the index of the target (Class)
# drop=True means "forget the old row numbers completely"
c_target_testing = c_target_testing.reset_index(drop=True)
# 2. OPTIONAL: You might need to do this for features too if they aren't already 0..n
# (It's safe to run this even if they are already reset)
c_features_testing = c_features_testing.reset_index(drop=True)
test_datasetv1=pd.concat([c_features_testing,c_target_testing],axis=1)
#saving the new test dataset
test_datasetv1.to_csv("Test_Datasetv1.csv",index=False)
print("test dataset created")

```

```python
#smote:synthetic minority resampling technique

smotev1=SMOTE(random_state=42)
#checking balance before applying smote
print("dataset balance before smote")
print(c_target_training.value_counts())
#applying smote on training dataset
c_features_training_resampled,c_target_training_resampled=smotev1.fit_resample(c_features_training,c_target_training)
#balance after applying smote
print("smote applied")
print(c_target_training_resampled.value_counts())


```

model training

```python
#logistic regression

#creating the model
lr_model=LogisticRegression(max_iter=5000,random_state=42)
print("training the model")
lr_model.fit(c_features_training_resampled,c_target_training_resampled)
print("training complete")
```

```python
#model predictions
lr_predictions=lr_model.predict(c_features_testing)
```

```python
#model evaluation
print("\n ---model evaluation report--- \n")
print(classification_report(c_target_testing,lr_predictions,target_names=['Not Fraud (0)', 'Fraud (1)']))
print("\n--- Logistic Regression Confusion Matrix ---")
print(confusion_matrix(c_target_testing, lr_predictions))
```

Logistic Regression Model Summary:
Accuracy (99%): This score is high but misleading because the data is imbalanced.
Fraud Recall (85%): This is the key metric. It means we successfully caught 85% of all actual frauds.
Fraud Precision (15%): This is the trade-off. It means that when the model did predict 'Fraud', it was only correct 15% of the time, resulting in many false alarms.
Conclusion: This model is very sensitive (catches most frauds) but not very precise (creates a lot of false positives).

```python
#random forest classifier with grid search
##THE MODEL TO OPTIMIZE THE F1 SCORE
#USING class_weight = balanced(the balanced one)

print("Starting the  'balanced' model search...")
print("This will find the best 'F1-Score' (balance of Precision and Recall).")

model = RandomForestClassifier(random_state=42, n_jobs=-1)

# A smaller, smarter grid to test.
# We are letting the grid *test* if 'class_weight' is good or not.
param_grid = {
    'n_estimators': [100, 150],  # Kept it small
    'max_depth': [20, 30],       # Testing deeper trees
    'class_weight': ['balanced', None] # Let the grid decide the best balance
    #'class_weight':['balanced']#just keeping the balanced settings for better f1 
}

# --- THIS IS THE KEY ---
# We are optimizing for the 'f1' score, which is the
# perfect balance between Precision and Recall.
grid_search = GridSearchCV(
    estimator=model, 
    param_grid=param_grid, 
    cv=5,                  # Use 3 folds instead of 5 to make it faster
    scoring='recall',          # <-- THE GOAL: "Be a smart, balanced guard"
    n_jobs=-1,             # Use all your computer's power
    verbose=2              # This will show you progress!
)

# Fit the grid search
grid_search.fit(c_features_training, c_target_training) # Use your train variables

# Get the one, true, balanced model
best_balanced_model = grid_search.best_estimator_

print("---")
print(f"Search complete! The best balanced model is:")
print(grid_search.best_params_)

# Save this new, final model
#joblib.dump(best_balanced_model, 'fraud_model.joblib')

print("---")
print("training complete with best fitting")
```

the grid search with scoring parameter f1 showcased optimization towards using model with class weight =none ,which provided great precision but lower recall scores, while the model was good as project in real world scenarios a fraud model is more optimized towards higher recall score as catching fraudster is more important than flagging customers,
for this we rerun the grid search of random forest model but with scoring parameter set to recall , which will fetch more optimized recall rates but lower precision rates.

```python
#comparision between balanced ,none and smote

def comparision():
    print("comparing the three models")
print("-"*50)
print("model 1: CLASS WEIGTHS=NONE")
print("-"*50)
print("\n training model with no weigths...")
#defining parameters

#model 1

model_1= RandomForestClassifier(
    n_estimators=150,#using parameters suggested by grid search
    max_depth=20,
    random_state=42,
    n_jobs=-1,
    class_weight=None #no weights
)
#training the model
model_1.fit(c_features_training,c_target_training)
#evaluating the model
print("evaluating model 1...")
pred_1=model_1.predict(c_features_testing)
print(classification_report(c_target_testing,pred_1,target_names=['Normal','Fraud']))

#model 2

print("-"*50)
print("model 2:CLASS WEIGHTS=BALANCED")
print("-"*50)
print("\n training model with balanced weigths...")
#defining parameteres
model_2=RandomForestClassifier(
    n_estimators=150,
    max_depth=20,
    random_state=42,
    n_jobs=-1,
    class_weight="balanced"
)
#training
model_2.fit(c_features_training,c_target_training)
#evaluating
print("evaluating model 2...")
pred_2=model_2.predict(c_features_testing)
print(classification_report(c_target_testing,pred_2,target_names=['Normal','Fraud']))

#model 3

print("-"*50)
print("model 3: SMOTE DATA")
print("-"*50)
print("\n training model with smote data...")
#def parameters
model_3=RandomForestClassifier(
    n_estimators=150,
    n_jobs=-1,
    max_depth=20,
    random_state=42,
    class_weight=None  #the smote data is already balanced
)
#training
print("using the resampled data")
model_3.fit(c_features_training_resampled,c_target_training_resampled)
#evaluation
print("evaluating model 3...")
pred_3=model_3.predict(c_features_testing)
print(classification_report(c_target_testing,pred_3,target_names=['Normal','Fraud']))
if __name__ == "__main__":
    comparision()
```

analysis of the comparision: 
SMOTE
the precision dropped with smote data because smote creates synthetic fraud cases ,which blurs the line between fraud and real cases.
NONE AND BALANCED
the balanced model , balances the class weight by paying more attention and cost to the fraud cases where each fraud is given 500 weight,this causes the model to overfit. By weighing the training fraud so heavily, it overfit and lost the ability to generalize to the test set.

```python
#using model with no class weights for overall better precision and recall
print("finally using the model with no class weigths...")
print("using fresh new model to avoid confusion and misfits ")
final_model_rf=RandomForestClassifier(
    n_estimators=150,
    n_jobs=-1,
    class_weight=None,
    random_state=42,
    max_depth=20
)

#training

print("training the model with the training data")
final_model_rf.fit(c_features_training,c_target_training)

print("final model testing")
final_predictions=final_model_rf.predict(c_features_testing)

#report
print("final training report")
print(classification_report(c_target_testing,final_predictions,target_names=['Fraud','Normal']))
print("confusion matrix for final model")
confusion_matrix(c_target_testing,final_predictions)
```

analysis of the prediction report:
the no class weight model , is optimized more for precision and less for recall, when it does tell fraud ,it is right for 97 percent of times

while this model works fine,the recall rate is lower and the model is more optimized towards precision while in real world fraud detections the bank prefers higher recall to catch fraudster even if it bothers some customers, for this we will use model optimized for recall rates


```python
#using higher recall model as final model
print("using recall priority model")
recall_model=RandomForestClassifier(
    n_estimators=100,
    max_depth=20,
    n_jobs=-1,
    class_weight=None,
  
    random_state=42

)
print("training the recall model...")
recall_model.fit(c_features_training,c_target_training)

print("testing recall model")
recall_predictions=recall_model.predict(c_features_testing)

print("final report")
print(classification_report(c_target_testing,recall_predictions,target_names=['Normal','Fraud']))

print("confusion_matrix")
confusion_matrix(c_target_testing,recall_predictions)
```

using this model instead of using smote rf , because we can control recall by adding a threshold bar without loosing on precision.

```python
#saving and running a simulation 

joblib.dump(recall_model,"recall_model.joblib")
print("finall prediciton model saved to recall_model...")
```

```python
#running a test simulation
print("\n .....TEST SIMULATION.....")
#selecting a random fraud case from the test dataset
fraud_case=np.where(c_target_testing==1)[0]
if len(fraud_case) > 0:
    #  Use np.random.choice to pick a RANDOM index 
    random_idx = np.random.choice(fraud_case)
    #selecting the first fraud transaction encountered
    sim_data=c_features_testing.iloc[random_idx].values.reshape(1,-1)  #iloc method finds the data for the case selected
    
# Get the prediction
    prob = recall_model.predict_proba(sim_data)[0][1]
    pred = recall_model.predict(sim_data)[0]
    
    status = "🚨 FRAUD" if pred == 1 else "✅ SAFE"
    print(f"Simulating Transaction #{random_idx}...")
    print(f"Prediction: {status} (Confidence: {prob:.2%})")
    
    if pred == 1:
        print("RESULT: Success! The model caught the thief.")
    else:
        print("RESULT: Missed. (Try lowering the threshold in the dashboard).")
```
